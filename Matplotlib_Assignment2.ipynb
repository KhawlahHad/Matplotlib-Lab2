# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 5, 8, 13, 20]

plt.figure(figsize=(8, 6)) 

plt.plot(x, y, marker='o', linestyle='-')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Chart')

plt.show()
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python

plt.figure(figsize=(8, 6))  

plt.plot(x, y, marker='o', linestyle='--', markersize=10, linewidth=3, label='Fast')
plt.plot(x, x, marker='o', linestyle='--', markersize=10, linewidth=3, label='Normal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Chart with Provided Data')
plt.legend()  

plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))  

axs[0].plot(x_plot1, y_plot1, marker='o', linestyle='-')
axs[0].set_xlabel('X-Axis')
axs[0].set_ylabel('Y-Axis')
axs[0].set_title('First Chart')

axs[1].plot(x_plot2, y_plot2, marker='o', linestyle='--')
axs[1].set_xlabel('X-Axis')
axs[1].set_ylabel('Y-Axis')
axs[1].set_title('Second Chart')

axs[2].plot(x_plot3, y_plot3, marker='o', linestyle='--')
axs[2].set_xlabel('X-Axis')
axs[2].set_ylabel('Y-Axis')
axs[2].set_title('Third Chart')

plt.suptitle('My Data Visualization Assignment', fontsize=16)

plt.tight_layout()

plt.show()
```

# Q5: Define an array that has a values of your top three TV series and define another array to hold the rating of the three TV series out of 10. Create a bar chart using the two arrays
- Set a new color for the bars.
- Change the width of the bars to width 0.5.

```python
tv_series = ["Black Mirror", "Game of Thrones", "Veep"]
ratings = [9.5, 9.3, 8.7]

bar_colors = ['violet', 'navy', 'pink']
plt.figure(figsize=(8, 6))

plt.bar(tv_series, ratings, color=bar_colors, width=0.5)

plt.xlabel('TV Series')
plt.ylabel('Rating')

plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
carAge = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
carSpeed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78]

point_colors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan', 'magenta']

plt.figure(figsize=(8, 6)) 

plt.scatter(carAge, carSpeed, c=point_colors, s=350, alpha=0.4)

plt.xlabel('Car Age (years)')
plt.ylabel('Car Speed (mph)')
plt.title('Car Age vs. Car Speed')

plt.legend()

plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
python_libraries = ["NumPy", "Pandas", "Matplotlib", "Seaborn", "Plotly", "Scikit-learn"]
skills = [9, 8, 7, 6, 5, 4]

wedge_colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'lightseagreen']
explode = (0.3, 0, 0, 0, 0, 0)
plt.figure(figsize=(8, 8)) 

plt.pie(skills, labels=python_libraries, colors=wedge_colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)

plt.axis('equal')

plt.title('My Python Library Skills')

plt.show()
```

# Q8: Create a histogram with 250 values and 15 bars

```python

data = np.random.randn(250)

plt.figure(figsize=(8, 6))  
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, color='lightpink')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```
